```python
# We are going to clean the names before isolating the ones we are interested in
df.columns = df.columns.str.capitalize().str.replace('(', '').str.replace(')', '').str.replace('y/n', '').str.strip()
df.columns
```

```python
# We eliminate rows that don't contain enough data to work with setting thresh to a number that contain a minimum
# amount of columns with information
# Column names have whitespace. We can clean that up with .strip()
# We will also put everything in lowercase

df_columns = df[["Date", "Year", "Time", "Type", "Country", "Area", "Activity", "Age", "Sex", "Fatal", "Species"]]
df_columns
```

```python
# Ways to isolate rows and columns
df.loc[[2,0],["Date","Time"]]
```

```python
# Checking data types
df_columns.dtypes 
```

```python
# We want to eliminated rows that don't contain sufficient information. 
# Here, We have decided to elminate rows that don't fill at least 6 columns
df_row_col = df_columns.dropna(thresh=6, axis=0)
df_row_col
```

```python
df_row_col.isnull().sum()
```

```python
df_row_col.isnull().sum().sum()
```

```python
# We will try to find unexpected missing values in certain columns

cnt=0
for row in df_row_col['Area']:
    try:
        int(row)
        df.loc[cnt, 'Area']=np.nan
    except ValueError:
        pass
    cnt+=1
```

```python
# This converts all Nan values to "n/a" so we can work with those values more effectively.
df_na = df_row_col.fillna("")
df_na
```

```python
print(df_row_col['Species'].unique())
```

```python
# Using regex we want to extract the type of shark only. We expect to find the type immediately before the word shark.
# We also need a list of non_valid words, until we come up with a more efficient way to handle this issue.

non_valid = ["m", "a","two", "unidentified", "foot","metre","several", "large", "larger", "another", "little", "no", "but", "for","small", "of", "lb", "kg", "as", "the", "from", "same"]
species_extr = (df_row_col['Species']
                .str.extract(r'([a-zA-Z]+)\s[S-s]hark', expand=False)
                .str.lower()
                .fillna("--")
                .replace(non_valid,"--"))


```

```python
species_extr
```

```python
clean_species = species_extr.to_frame("Species")
```

```python
# This confirms that .to_frame converted the series to a dataframe
# We can find a way to join this to the table or start appending clean columns to it.
type(clean_species)
```
